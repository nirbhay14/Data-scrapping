```python
import selenium
```

```python
from selenium import webdriver
PATH = "C:\Program Files (x86)/chromedriver.exe"
driver = webdriver.Chrome(PATH)
driver.get("https://www.giiresearch.com/material_report.shtml")
title=[]
published=[]
code=[]
published_date=[]
page=[]
price=[]

matches = driver.find_elements_by_tag_name('tr')
for match in matches:
    titles = match.find_elements_by_xpath("//td//div[@class='plist_t_box']")
    publishers = match.find_elements_by_xpath("//div[@class='plist_pubinfo']//div[@class='plist_info_dd2']")
    codes = match.find_elements_by_xpath("//div[@class='plist_codeinfo']//div[@class='plist_info_dd2']")
    dates=match.find_elements_by_xpath("//div[@class='plist_dateinfo']//div[@class='plist_info_dd2']")
    pages = match.find_elements_by_xpath("//div[@class='plist_pageinfo']//div[@class='plist_info_dd2']")
    prices = match.find_elements_by_xpath("//div//span[@class='price_usd']")



for tt in titles:
    title.append(tt.text)
for pd in publishers:
    published.append(pd.text)
for cd in codes:
    code.append(cd.text)
for dd in dates:
    published_date.append(dd.text)
for pg in pages:
    page.append(pg.text)
for pr in prices:
    price.append(pr.text)    

driver.quit()
```

```python
import pandas as pd
```

```python
df = pd.DataFrame({'Title':title, 'Publishers':published, 'Product_code':code,'Published_date':published_date,'Pages':page,'Price':price})
```
